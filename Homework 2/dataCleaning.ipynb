```python
oct_data = monthly.loc[monthly['Month, Year'].apply(lambda x: x[:3]=='Oct')]
oct_data=oct_data.T
oct_data.iloc[1:]
```

```python
oct_index=list(oct_data.index)
oct_index_region = [x[-2:] for x in oct_index[1:]]
oct_index_item=[x[:2] for x in oct_index[1:]]
mapped_oct_index_items = [oct_index_item,oct_index_region]
mapped_oct_index = pd.MultiIndex.from_arrays(mapped_oct_index_items, names = ['item', 'region'])
mapped_oct_index_items
```

```python
october=pd.DataFrame(oct_data.iloc[1:].values,index=mapped_oct_index, columns=['oct_16', 'oct_17', 'oct_18'])
october
```

```python
october = october.groupby(['item','region']).max()
october
```
